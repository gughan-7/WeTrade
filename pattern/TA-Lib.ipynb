```python
def get_patterns(df):
    engulfing = talib.CDLENGULFING(df['Open'], df['High'], df['Low'], df['Close'])
    morning_star = talib.CDLMORNINGSTAR(df['Open'], df['High'], df['Low'], df['Close'])
    eveningstar = talib.CDLEVENINGSTAR(df['Open'], df['High'], df['Low'], df['Close'])
    whitesoldiers = talib.CDL3WHITESOLDIERS(df['Open'], df['High'], df['Low'], df['Close'])
    dojistar = talib.CDLDOJISTAR(df['Open'], df['High'], df['Low'], df['Close'])
    hammer = talib.CDLHAMMER(df['Open'], df['High'], df['Low'], df['Close'])
    harami = talib.CDLHARAMI(df['Open'], df['High'], df['Low'], df['Close'])
    invertedhammer = talib.CDLINVERTEDHAMMER(df['Open'], df['High'], df['Low'], df['Close'])
    shootingstar = talib.CDLSHOOTINGSTAR(df['Open'], df['High'], df['Low'], df['Close'])
    
    
    df['Engulfing'] = engulfing
    df['Morning Star'] = morning_star
    df['Evening Star'] = eveningstar
    df['3 White Soldiers'] = whitesoldiers
    df['Doji Star'] = dojistar
    df['Hammer'] = hammer
    df['Harami'] = harami
    df['Inverted Hammer'] = invertedhammer
    df['Shooting Star'] = shootingstar
    
    df = df
    return df
```
